```python
#Quitar links, menciones y hashtag

import re 
```

```python
pattern1 = '?P<pic>pic.twitter.com/[^\s]+' #patrones de regex
pattern2 = '?P<url>https?://[^\s]+' #patrones de URL

def text_clean(row): #regresa el resultado en columnas
    text = row['tweets'] #permite que el resultado se convierta en un objeto de pandas
    
    links = [tuple(j for j in i if j)[-1] for i in re.findall(f"({pattern1})|({pattern2})",text)]#busca palabras que contienen los patrones
    for link in links:
        text = text.replace(link,"")#quita las palabras que tienen esos patrones
    
    hashtags = [interaction for interaction in text.split() if interaction.startswith("#")]#guarda palabras que comienzan con #
    for hashtag in hashtags:
        text = text.replace(hashtag, "")#quita los hashtag
    
    mentions = [interaction for interaction in text.split() if interaction.startswith('@')]#guarda palabras que comienzan con @
    for mention in mentions:
        text = text.replace(mention, "")#quita menciones
        
    return text, links, hashtags, mentions
```

```python
datos[['texto_limpio', 'links', 'hashtags', 'mentions']] = datos.apply(text_clean, axis=1, result_type='expand')
datos
```

```python
import regex
import emoji
```

```python
def get_emojis(text):
    emoji_list = []
    datos = regex.findall(r'\X', text) #busca caracteres que empiecen con \X(emojis)
    for word in datos:
        if any(char in emoji.UNICODE_EMOJI['en'] for char in word):
            emoji_list.append(word) #agrega una lista de emojis a la base

    return emoji_list
```

```python
datos['emojis'] = datos['texto_limpio'].apply(lambda text: get_emojis(text))
datos.head()
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
stopwords_sp = stopwords.words('spanish')
```

```python
#Vectorización

def pre_procesado(texto):
    texto = texto.lower()                                                         #pasar a minúsculas
    texto = re.sub(r"[\W\d_´]+", " ", texto)                                      #eliminar caracteres
    texto = " ".join([palabra for palabra in texto.split() if palabra not in stopwords_sp]) #quitar palabras vacías
    return (texto)

tfidf_vect = TfidfVectorizer(preprocessor=pre_procesado)
tfidf = tfidf_vect.fit_transform(datos.texto_limpio.values)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
voc = [k for k,v in sorted(tfidf_vect.vocabulary_.items(), key=lambda kv: kv[1])]
temp = pd.DataFrame(tfidf.toarray())
temp = cosine_similarity(temp.values)
temp = pd.DataFrame(temp)
temp
```

```python
import scipy.cluster.hierarchy as sch
import numpy as np
```

```python
pairwise_distances = sch.distance.pdist(temp)
linkage = sch.linkage(pairwise_distances, method='ward')
idx_to_cluster_array = sch.fcluster(linkage, pairwise_distances.max() * 0.5, criterion='distance')
idx = np.argsort(idx_to_cluster_array)
temp = temp.copy()
    
temp2 = temp.iloc[idx, :].T.iloc[idx, :]
my_idx = idx_to_cluster_array

temp2.index = ['tweet #'+str(i) for i in temp2.columns]
temp2.columns = temp2.index
```

```python
#Gráfico

trace = go.Heatmap(z=temp2.values.tolist(),
                   x=temp2.index.values,
                   y=temp2.columns.values,
                   colorscale='Oranges')

layout = go.Layout(title='Mapa de Calor: Tweets 2011-2012 en Chile, Keyword "Ir a Chile"',
                   width=650, height=600)

fig = go.Figure(data=[trace],layout=layout)
iplot(fig)
```
